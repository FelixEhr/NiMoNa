## Task 1: Plotting and Functions

```python
#x-Wertebereich
xrange=np.linspace(-4,4,1000)
```

```python
def f(x,a):
    f=a*np.sin(2*np.pi*x)
    return f
```

```python
def g(x,b):
    g=1+b*x**2
    return g
```

```python
def h(x,c):
    h=f(x,g(x,c))
    return h
```

```python
c=([0,0.5,1.0])
Farbe=["orangered","darkgreen","darkred"]
Name=["c=0","c=0.5","c=1.0"]
Ebene=([5,4,3])

plt.grid(True, zorder=1)

for i in range(len(c)):
    plt.plot(xrange,h(xrange,c[i]), zorder=Ebene[i], linewidth=2.5, color=Farbe[i], linestyle="-", label=Name[i])

plt.xlim(-4,4)
plt.xlabel("x")
plt.ylim(-16,16)
plt.ylabel("y")

plt.legend(shadow=True, loc=9)
```

```python
plt.grid(True, zorder=4)

a=1

plt.plot(xrange,f(xrange,a), zorder=5, linewidth=3, color="darkred", label="f(x,1)")
plt.plot(xrange,g(xrange,a), zorder=5, linewidth=3, color="darkgreen", label="g(x,1)")
plt.plot(xrange,h(xrange,a), zorder=5, linewidth=2.5, color="orangered", linestyle="--", label="h(x,1)")

plt.xlim(-4,4)
plt.xlabel("x")
plt.ylim(-16,16)
plt.ylabel("y")

plt.legend(shadow=True, loc=9)
```
